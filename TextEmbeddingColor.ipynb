```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['text.usetex'] = False

import string
from TextEmbedding import WordEmbedding
```

```python
we = WordEmbedding()
```

```python
# Reshape the embeddings into a list of 2D matrices
embedding_images = we.embedding_chars
# embedding_images -= embedding_images.min()
# embedding_images /= embedding_images.max()
# embedding_images -= embedding_images.mean()
# embedding_images /= embedding_images.std()
# embedding_images = embedding_images * 0.5 + 0.5
embedding_images -= embedding_images.mean()
embedding_images /= embedding_images.max()
embedding_images = embedding_images * 0.5 + 0.5
tokens = list(string.printable)
tokens[-6] = " "
tokens[-5] = "\\n"
tokens[-4] = "\\t"
tokens[-3] = "\\r"
tokens[-2] = "\\v"
tokens[-1] = "\\f"
```

```python
# Plot the embedding images using Matplotlib
fig, axs = plt.subplots(nrows=10, ncols=10, figsize=(20, 20))
for i, token in enumerate(tokens):
    axs[i//10, i % 10].imshow(embedding_images[i].clip(0, 1).reshape(32, 32),
                              cmap="PiYG", interpolation='nearest')
    axs[i//10, i % 10].axis('off')
    axs[i//10, i % 10].axis('tight')
    axs[i//10, i % 10].set_title(token)
plt.show()
```
